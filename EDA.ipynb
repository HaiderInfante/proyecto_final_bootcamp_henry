```python
import pandas as pd
```

```python
archivo_csv = '2c4fe398-9e6d-47c7-b2ea-9640a72cd7ed_Data.csv'

# Leer el archivo CSV y crear un DataFrame
df = pd.read_csv(archivo_csv)
```

```python
df.head()
```

```python
# Supongamos que tienes un DataFrame llamado 'df' con la estructura proporcionada

# Lista de columnas que no se deben incluir en la fusión
columnas_no_a_melt = ['Country Name', 'Country Code', 'Series Name', 'Series Code']

# Utiliza pd.melt para combinar las columnas de años en una sola columna
df_melted = pd.melt(df, id_vars=columnas_no_a_melt, var_name='Año', value_name='Valor')

# Esto creará un nuevo DataFrame 'df_melted' con las columnas 'Country Name', 'Country Code', 'Series Name', 'Series Code', 'Año' y 'Valor'
```

```python
df_melted
```

```python
# Elimina las filas que contienen '..' en la columna 'Valor'
df_final = df_melted[df_melted['Valor'] != '..']

# Agrupa por 'Series Name' y cuenta la cantidad de filas con valores en la columna 'Valor'
conteo_series_name = df_final.groupby('Series Name')['Valor'].count().reset_index()

# Renombra la columna de conteo para mayor claridad
conteo_series_name.rename(columns={'Valor': 'Cantidad'}, inplace=True)

# Ahora 'conteo_series_name' contiene el conteo de "Series Name" con valores asociados
# Puedes imprimirlo o trabajar con él según tus necesidades
conteo_series_name = conteo_series_name.sort_values(by='Cantidad', ascending=False)
conteo_series_name
```

```python
# Filtra las filas donde 'Cantidad' sea igual a 132
result = conteo_series_name[conteo_series_name['Cantidad'] == 128]

# Imprime o trabaja con el DataFrame 'result', que contendrá solo las filas donde 'Cantidad' es igual a 132
result
```

```python
# Supongamos que tienes un DataFrame llamado 'df' que deseas exportar a un archivo CSV.

# Especifica el nombre del archivo CSV de destino
nombre_archivo_csv = 'World Development Indicators.csv'

# Utiliza la función to_csv para exportar el DataFrame a CSV
result.to_csv(nombre_archivo_csv, index=False)  # El parámetro index=False evita que se incluyan los índices en el archivo CSV
```
