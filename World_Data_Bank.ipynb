```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Archivo borrador. Fuente WB
df_WB= pd.read_excel("./Borrador_Data_WB.xlsx")
df_WB.head()
```

```python
# Para realizar correlaciones, primero excluyo las columnas string.
columnas_no_numericas = ['Country']
# Tomo las primeras 20 colunas numéricas, porque cuando haga el gráfico, queda ilegible.
columnas_numericas = df_WB.columns.difference(columnas_no_numericas)[:20]
# Calculo la matriz de correlación
correlaciones = df_WB[columnas_numericas].corr()
print(correlaciones)
```

```python
# Visualizo la correlación.
plt.figure(figsize=(12, 10))  
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Mapa de calor de correlación")
plt.show()
```
