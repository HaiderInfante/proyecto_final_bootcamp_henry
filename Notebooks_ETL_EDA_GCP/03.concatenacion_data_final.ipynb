```python
df_database_WorldBank_OMS = pd.read_csv('datasets_limpios/df_dababase_WorlBank_OMS.csv')
df_database_externa = pd.read_csv('datasets_limpios/df_database_externa.csv')
```

```python
# Supongamos que tienes tu DataFrame llamado df
# Y solo deseas mantener la columna "Gasto público social según la clasificación por funciones del gobierno (en moneda nacional a precios corrientes)"
columna_deseada = 'Gasto público social según la clasificación por funciones del gobierno (en moneda nacional a precios corrientes)'

# Eliminar filas con NaN o nulos en la columna deseada
df_gasto_publico = df_database_externa.dropna(subset=[columna_deseada])

columnas_a_eliminar = ['Prevalence of raised blood glucose/diabetes in adults (%)', 'Proportion of adults who are obese']

# Utiliza el método .drop() para eliminar las columnas
df_gasto_publico = df_gasto_publico.drop(columnas_a_eliminar, axis=1)
# El DataFrame ahora solo contiene filas sin NaN o nulos en la columna deseada
df_gasto_publico
```

```python
df_merged = df_database_WorldBank_OMS.merge(df_gasto_publico, on=['País', 'Año'], how='left')
df_merged
```

```python
# Supongamos que tienes tu DataFrame llamado df
# Y solo deseas mantener la columna "Gasto público social según la clasificación por funciones del gobierno (en moneda nacional a precios corrientes)"
columna_deseada = 'Prevalence of raised blood glucose/diabetes in adults (%)'

# Eliminar filas con NaN o nulos en la columna deseada
df_glucosa = df_database_externa.dropna(subset=[columna_deseada])

columnas_a_eliminar = ['Gasto público social según la clasificación por funciones del gobierno (en moneda nacional a precios corrientes)', 'Proportion of adults who are obese']

# Utiliza el método .drop() para eliminar las columnas
df_glucosa = df_glucosa.drop(columnas_a_eliminar, axis=1)
# El DataFrame ahora solo contiene filas sin NaN o nulos en la columna deseada
df_glucosa
```

```python
df_merged_2 = df_merged.merge(df_glucosa, on=['País', 'Año'], how='left')
df_merged_2
```

```python
# Supongamos que tienes tu DataFrame llamado df
# Y solo deseas mantener la columna "Gasto público social según la clasificación por funciones del gobierno (en moneda nacional a precios corrientes)"
columna_deseada = 'Proportion of adults who are obese'

# Eliminar filas con NaN o nulos en la columna deseada
df_obesidad = df_database_externa.dropna(subset=[columna_deseada])

columnas_a_eliminar = ['Gasto público social según la clasificación por funciones del gobierno (en moneda nacional a precios corrientes)', 'Prevalence of raised blood glucose/diabetes in adults (%)']

# Utiliza el método .drop() para eliminar las columnas
df_obesidad = df_obesidad.drop(columnas_a_eliminar, axis=1)
# El DataFrame ahora solo contiene filas sin NaN o nulos en la columna deseada
df_obesidad
```

```python
import pandas as pd

# Supongamos que tienes tu DataFrame llamado df
# Elimina las filas duplicadas basadas en las columnas 'País' y 'Año'
df_obesidad_sin_duplicados = df_obesidad.drop_duplicates(subset=['País', 'Año'])

# El DataFrame df_sin_duplicados ahora contiene solo las filas únicas en función de 'País' y 'Año'
df_obesidad_sin_duplicados
```

```python
df_merged_3 = df_merged_2.merge(df_obesidad_sin_duplicados, on=['País', 'Año'], how='left')
df_merged_3
```
