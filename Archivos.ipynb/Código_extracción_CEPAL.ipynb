```python
import requests
import pandas as pd

countries = ["ATG","ARG", "COL", "BRB"]
indicators = ["3126"]
start_year = "1990"
end_year = "2022"
data = []

base_url = "https://api-cepalstat.cepal.org/cepalstat/api/v1/indicator/"

for country_code in countries:
    for indicator in indicators:
        url = f"{base_url}/{indicator}/data?filter=Country eq '{country_code}' and Year ge {start_year} and Year le {end_year}"

        response = requests.get(url)

        if response.status_code == 200:
            result = response.json()
            if 'body' in result and 'data' in result['body']:
                for entry in result['body']['data']:
                    value = entry.get('value')
                    country = entry.get('iso3')
                    year = entry.get('Year')  
                    indicator_name = result['body']['metadata']['indicator_name']
                    data.append({"País": country, "Año": year, "Indicador": indicator_name, "Valor": value})

data_df = pd.DataFrame(data)
data_df
```

```python
import requests
import pandas as pd
import re

countries = ["ATG","ARG","BHS","BRB","BLZ","BOL","BRA","CAN","CHL","COL","CRI","CUB","DOM","ECU","GRD","GTM",
             "GUY","HTI","HND","JAM","SLV","USA","MEX","NIC","PAN","PRY","PER","DOM","KNA","LCA","VCT","SUR",
             "TTO","URY","VEN"]
indicators = ["3126"]
start_year = "1990"
end_year = "2022"
data = []

base_url = "https://api-cepalstat.cepal.org/cepalstat/api/v1/indicator"

for country_code in countries:
    for indicator in indicators:
        url = f"{base_url}/{indicator}/data?filter=Country eq '{country_code}' and Year ge {start_year} and Year le {end_year}"

        response = requests.get(url)

        if response.status_code == 200:
            result = response.json()
            if 'body' in result and 'data' in result['body']:
                # Extraer el año del filtro en la URL
                year_match = re.search(r'(\d{4})', url)
                if year_match:
                    start_year = int(year_match.group(1))
                    end_year = start_year

                for entry in result['body']['data']:
                    value = entry.get('value')
                    country = entry.get('iso3')
                    indicator_name = result['body']['metadata']['indicator_name']
                    data.append({"País": country, "Año": start_year, "Indicador": indicator_name, "Valor": value})

data_df = pd.DataFrame(data)
data_df
```
