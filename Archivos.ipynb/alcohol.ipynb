```python
import pandas as pd
import numpy as np
```

```python
# Fuente externa - World Health Organization
df_alcohol= pd.read_excel("./alcohol.xlsx")
df_alcohol.head()
```

```python
# Cambio nombres de algunas columnas en el df, para que coincida con el df de WB.
df_alcohol.rename(columns={"SpatialDim_es": "Country"}, inplace=True)
df_alcohol.rename(columns={"SpatialDim": "Country_Code"}, inplace=True)
df_alcohol.rename(columns={"TimeDim": "Year"}, inplace=True)
df_alcohol.head()
```

```python
# Procedo a dejar las columnas necesarias para hacer merge con el dataset de WB.
df_alcohol = df_alcohol.drop("paho_indicator_id", axis=1) 
df_alcohol = df_alcohol.drop("Country_Code", axis=1) 
df_alcohol = df_alcohol.drop("SpatialDimType", axis=1) 
df_alcohol = df_alcohol.drop("SpatialDim_en", axis=1) 
df_alcohol = df_alcohol.drop("data_source_specific", axis=1) 
df_alcohol = df_alcohol.drop("data_provider_specific", axis=1) 
df_alcohol = df_alcohol.drop("data_secondary_source", axis=1) 
df_alcohol = df_alcohol.drop("source_url", axis=1) 
df_alcohol.head()
```

```python
df_alcohol.to_csv('consumo_alcohol.csv', index=False)
```
