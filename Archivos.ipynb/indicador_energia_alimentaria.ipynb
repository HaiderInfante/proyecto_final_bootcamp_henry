```python
import pandas as pd
import numpy as np
```

```python
# Fuente externa - United Nations Statistics Division (UNSD),
df_alimentaria= pd.read_excel("./Energia_alimentaria - CEPAL.xlsx")
df_alimentaria.head()
```

```python
# Elimino las columnas no necesarias
df_alimentaria = df_alimentaria.drop("notes_ids", axis=1) 
df_alimentaria = df_alimentaria.drop("source_id", axis=1) 
df_alimentaria.head()
```

```python
orden_columnas = ['País__ESTANDAR', "Años__ESTANDAR",'Necesidades promedio de energía alimentaria kilocalorias']
df_alimentaria = df_alimentaria[orden_columnas]  
df_alimentaria
```

```python
# Cambio nombres de algunas columnas en el df, para que coincida con el df de WB.
df_alimentaria.rename(columns={"País__ESTANDAR": "Country"}, inplace=True)
df_alimentaria.rename(columns={"Años__ESTANDAR": "Year"}, inplace=True)
df_alimentaria
```
