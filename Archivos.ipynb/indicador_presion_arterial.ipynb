```python
import pandas as pd
import numpy as np
```

```python
# Fuente externa - World Health Organization
df_presion_arterial= pd.read_excel("./presión_ arterial.xlsx")
df_presion_arterial.head()
```

```python
df_presion_arterial.info()
```

```python
df_presion_arterial.describe()
```

```python
# Procedo a dejar las columnas necesarias para hacer merge con el dataset de WB.
df_presion_arterial = df_presion_arterial.drop("paho_indicator_id", axis=1) 
df_presion_arterial = df_presion_arterial.drop("SpatialDimType", axis=1) 
df_presion_arterial = df_presion_arterial.drop("SpatialDim_en", axis=1) 
df_presion_arterial = df_presion_arterial.drop("SpatialDim", axis=1)
df_presion_arterial = df_presion_arterial.drop("TimeDimType", axis=1) 
df_presion_arterial = df_presion_arterial.drop("data_source_specific", axis=1) 
df_presion_arterial = df_presion_arterial.drop("data_provider_specific", axis=1) 
df_presion_arterial = df_presion_arterial.drop("data_secondary_source", axis=1) 
df_presion_arterial = df_presion_arterial.drop("source_url", axis=1) 
df_presion_arterial.head()
```

```python
# Cambio nombres de algunas columnas en el df, para que coincida con el df de WB.
df_presion_arterial.rename(columns={"SpatialDim_es": "Country"}, inplace=True)
df_presion_arterial.rename(columns={"TimeDim": "Año"}, inplace=True)
df_presion_arterial.head()
```

```python
df_presion_arterial.to_csv("presion_arterial.csv", index=False)
```
